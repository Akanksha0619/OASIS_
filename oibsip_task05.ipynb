#### PROBLEM STATEMENTE 

Sales prediction means predicting how much of a product peuple will buy based on factors such as the amount you spend to advertise your product, the segment of people you advertise for, or the platform you are advertising on about your product.

Typically, a product and service-based business always need their Data Scientist to predict their future sales with every step they take to manipulate the cost of advertising their product. So lets start the task of sales prediction with machine learning using Python.

Import Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

Import Data

```python
data=pd.read_csv('Advertising.csv')
```

Plot Sales

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=data['TV'],y=data['Sales'])
plt.show()
plt.figure(figsize=(8,5))
sns.scatterplot(x=data['Radio'],y=data['Sales'])
plt.show()
plt.figure(figsize=(8,5))
sns.scatterplot(x=data['Newspaper'],y=data['Sales'])
plt.show()
```

```python
train,test = train_test_split(data,test_size=0.30)
```

```python
train_x=train[['TV', 'Radio', 'Newspaper']]
train_y=train.Sales

test_x=train[['TV', 'Radio', 'Newspaper']]
test_y=train.Sales
```

```python
train_x.head
```

```python
train_y.head
```

train model

```python
model= linear_model.LinearRegression()
model.fit(train_x,train_y)
predict=model.predict(test_x)
metrics.r2_score(predict,test_y)
```

```python
plt.figure(figsize=(10,5))
sns.regplot(x=test_y,y=predict,scatter_kws={'color':'red'})
```
